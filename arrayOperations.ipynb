<h3>Deep copy and Shallow copy</h3>

```python
A=np.array([1,2,3,4,5])
```

```python
B=A
```

```python
B[0]=12
```

```python
A
```

```python
B #Changing with B reflect changes in A its a shallow copy
```

```python
A=np.array([1,2,3,4,5])
```

```python
B=np.copy(A)
```

```python
B[0]=12
```

```python
B
```

```python
A # A array element are same this is deep copy
```

<h3>Matrix operations </h3>

```python
matx1=np.arange(1,5).reshape(2,2)
matx1
```

```python
matx2=np.arange(1,5).reshape(2,2)
matx2
```

```python
matx1@matx2
```

```python
np.dot(matx1,matx2)
```

```python
matx1.T
```

<h3>Array Stacking</h3>

Array stacking doest not work on 2D Matrix

```python
arr1=np.arange(1,11)
arr2=np.arange(11,21)
```

```python
np.hstack((arr1,arr2))
```

```python
np.vstack((arr1,arr2))
```

```python
np.column_stack((arr1,arr2))
```

<h3>Array Splitting</h3>

```python
myArr=np.arange(1,61).reshape(10,6)
print(myArr)
```

```python
print(np.hsplit(myArr,2))
```

```python
print(np.vsplit(myArr,5))
```
