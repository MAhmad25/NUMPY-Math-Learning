NUMPY Indexing and Slicing of Vector-1D

```python
import numpy as np
```

```python
arr=np.arange(11)
print(arr)
```

```python
print(arr[::3])
```

NUMPY Indexing and Slicing of Matrix-2D

```python
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
print(arr[3,3])
```

```python
arr[3:,3:]
```

Bool Indexing

```python
bool_index=arr %2==0
bool_index
```

```python
EvenValue=arr[bool_index]
print(EvenValue)
```
